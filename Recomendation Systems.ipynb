```python
import numpy as np
import pandas as pd
```

```python
column_names = ['user_id','item_id','rating','timestamp']
df = pd.read_csv('users.data',sep='\t',names=column_names)
```

```python
df.head()
```

```python
len(df)#kac kayıt oldugu 
```

```python
movie_titles = pd.read_csv("movie_id_titles.csv")
movie_titles.head()
```

```python
len(movie_titles)
```

```python
df = pd.merge(df, movie_titles, on = 'item_id')
df.head()
```

# Recommendation Sistemimizi Kuruyoruz

```python
moviemat = df.pivot_table(index='user_id',columns = 'title',values = 'rating')
moviemat.head()
```

## Amacımız Starwars filmine benzer film önerileri yapmak 

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
starwars_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
```

```python
similar_to_starwars
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns = ['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

Star Wars'a en yakın filmlerin korelasyonları aşağıdaki gibidir

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

Fakat elde ettiğimiz bu filmler biraz saçma çünkü içerisinde sadece 2 oy kullananlar olabilir
oyüzden 100 den az olan filmleri elememiz gerekiyor

```python
df.drop(['timestamp'], axis = 1)
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())

ratings.sort_values('rating',ascending=False).head()
```

```python
ratings['rating_oy_sayisi'] = pd.DataFrame(df.groupby('title')['rating'].count())#Her filmin aldıgı oy sayısı
ratings.head()
```

```python
#En çok oy alan filmleri büyükten küçüğe sıralayıp bakalım
ratings.sort_values('rating_oy_sayisi',ascending = False).head()
```

```python
corr_starwars.head()
```

```python
corr_starwars = corr_starwars.join(ratings['rating_oy_sayisi'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['rating_oy_sayisi']>100].sort_values('Correlation',ascending = False).head()
```
